## 2.Examine the data by using visualizations

### surface area vs population

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Region',y='Surface area (km2)',data=data)
plt.title('plot of region vs surface area',fontsize=15)
plt.xlabel('region',fontsize=12)
plt.ylabel('surface area',fontsize=12)
plt.xticks(rotation=90)
```

## Region vs International trade: Imports (million US$)

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Region',y='International trade: Imports (million US$)',data=data)
plt.title('plot of region vs International trade: Imports (million US$)',fontsize=15)
plt.xticks(rotation=90)
plt.tight_layout()
```

# 5. Region with more consistency with respect to GDP per capita (current US$)

```python
To find the standard deviation of GDP per capita (current US$) of different regions
```

```python
reg = data.groupby('Region')['GDP per capita (current US$)'].std()
reg
```

```python
A region has more consistency if the standard deviation with respect to GDP per capita (current US$) of that region is the least.
Now arranging the regions in the ascending order of standard deviation.
```

```python
reg.nsmallest()
```

```python
Thus Western Africa is more consistent with respect to  GDP per capita (current US$).
```
