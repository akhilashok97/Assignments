## 2. number of female senior citizen in the dataset

```python
seniorfemale=((df['gender']=='Female')&(df['SeniorCitizen']==1)).sum()
```

```python
print('no of female senior citizen :', seniorfemale)
```

```python
notsenior=((df['gender']=='Female')&(df['SeniorCitizen']==0)).sum()
```

```python
notsenior
```

```python
x=[notsenior,seniorfemale]
y=['notseniorcitizen','seniorcitizen']
plt.figure()
plt.pie(x,labels=y,autopct='%2.1f%%')
plt.title("Female citizen classification")
plt.show
```

### insight: here 16.3% of the total female are senior citizen

## 3.compare tenure with total charges

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],downcast='float',errors='coerce')
df.dtypes
```

```python
df.info()
```

```python
df['TotalCharges'].fillna(df['TotalCharges'].mean(),inplace=True)
df.info()
```

### box plot of Tenure vs total charges

```python
plt.figure(figsize=(15,10))
sns.barplot(x='tenure',y='TotalCharges',data=df)
plt.title('Plot of tenure vs total charges',fontsize=18)
```

### insight: here we could see that total charge is increasing with respect to tenure

## 4. contract preffered by senior citizen

```python
df.loc[df['SeniorCitizen']==1,['Contract']].value_counts().to_frame()
```

```python
fig=plt.figure(figsize=(8,8))
labels=["Month-to-Month","One Year","Two Year"]
plt.pie(df.loc[df["SeniorCitizen"]==1,["Contract"]].value_counts(),autopct= "%1.2f%%",labels=labels)
plt.legend(loc=(1.05,0))
plt.title("Contract preffered by senior citizens",fontsize=20)
plt.show()
```

#### insight: here month to month is preffered by most senior citizen (70.67%), followed by one year contract(16.64%) and then comes two year contract (12.70%)
