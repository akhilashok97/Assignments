# step 6: creating meaningful visualizations using the dataset

## 1.count plot of Iris-setosa,versicolor,virginica

```python
sns.countplot(data['Classification'])
plt.grid()
plt.title('Distribution of Iris Flowers',fontsize=15)
plt.xlabel('classification',fontsize=12)
plt.ylabel('count',fontsize=12)
```

**Here a count plot is obtained to visualize the distribution of Iris flowers.**

## 2.pairplot

```python
sns.pairplot(data)
```

**Here a pairplot is visualized which shows a representation of all the datas**

## 3.Box plot

```python
sns.boxplot(x='Classification',y='SL',data=data)
plt.title('Distribution of Iris Flowers based on SL',fontsize=15)
plt.xlabel('classification',fontsize=12)
plt.ylabel('SL',fontsize=12)
```

**Here a Box plot is obtained which depicts the distribution of iris flowers based on 'SL'**

## 4.Strip plot

```python
sns.stripplot(x="Classification",y="SW",data=data)
plt.title('Distribution of Iris Flowers based on SW',fontsize=15)
plt.xlabel('classification',fontsize=12)
plt.ylabel('SW',fontsize=12)
```

**Here the distribution of iris flowers on the basis of 'SW' is shown using a strip plot.**

## 5.Swarm plot

```python
sns.swarmplot(x="Classification",y="PL",data=data)
plt.title('Distribution of Iris Flowers based on PL',fontsize=15)
plt.xlabel('classification',fontsize=12)
plt.ylabel('PL',fontsize=12)
```

**Here using swarmplots, the distribution iris flowers on the basis of 'PL' is shown.**

## 6.Violin plot

```python
sns.violinplot(x="Classification",y="PW",data=data)
plt.title('Distribution of Iris Flowers based on PW',fontsize=15)
plt.xlabel('classification',fontsize=12)
plt.ylabel('PW',fontsize=12)
```

**Here the distribution of iris flowers on the basis of 'PW' is depicted using violin plot.**
