# Case study 1- Presentation of Data

## step 1 : Importing the libraries needed

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

## step 2 : Reading the dataset to the python environment

```python
data=pd.read_csv('cars_data.csv')
```

### displaying the contents of the dataset

```python
data
```

**Here a dataset is provided with details about cars,their make, model, selling price , buyer details etc**

## step 3: checking for null values present in the dataset

```python
data.isna()
```

**Here isna() function checks for null values in data set and gives a true or false accordingly**

**Inorder to get the count of the null values present in each column we are using a sum() along with isna()**

```python
data.isna().sum()
```

**From the above information using the sum() function it is clear that the dataset doesnot contain any null values**

## step 4: plotting a bar graph of male vs female buyers 

```python
plt.figure(figsize=(10,10))
sns.countplot(data['Buyer Gender'])
plt.grid()
plt.title('plot of male vs female buyers participated in the sale',fontsize=15)
plt.xlabel('Buyer Gender',fontsize=12)
plt.ylabel('count',fontsize=12)
```

**Inorder to show the count of male and female buyers, we are using value_counts() function**

```python
data['Buyer Gender'].value_counts()
```

**There are 5052 female buyers and 4948 male buyers as per the data set**

## step 5 : Finding the top 5 cars based on their sales price

### First step is to create a dataframe with columns having details about make,model and sale price of cars

```python
a=pd.DataFrame(data,columns=['Make','Model','Sale Price'])
```

```python
a
```

**The above data frame consist of car details with sale price not sorted. Inorder to sort the data we are doing the following step**

```python
b=a.sort_values('Sale Price',ascending=False)
```

```python
b
```

**Now with above data being sorted to our requirement we are using the head() function to display only the required number of cars**

```python
b.head(5)
```

**The top 5 cars are : volkswagen jetta III,
                       Audi A6,
                       Lexus LX,
                       Mazda MX-6,
                       Lexus ES**
